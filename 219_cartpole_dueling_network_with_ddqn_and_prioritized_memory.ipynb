```python
%matplotlib inline
from IPython.display import display, HTML
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import gym
from collections import deque
import random
import math
from copy import deepcopy

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
RANDOM_SEED = 0
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
torch.use_deterministic_algorithms(True)
```

```python
def transitions_to_tensors(transitions):
    state_list, action_list, reward_list, next_state_list, running_mask_list = [], [], [], [], []

    for transition in transitions:
        state, action, reward, next_state, running_mask = transition
        state_list.append(state)
        action_list.append([action])
        reward_list.append([reward])
        next_state_list.append(next_state)
        running_mask_list.append([running_mask])

    return (
        torch.tensor(state_list, dtype=torch.float),
        torch.tensor(action_list),
        torch.tensor(reward_list),
        torch.tensor(next_state_list, dtype=torch.float),
        torch.tensor(running_mask_list),
    )
    
class ReplayBuffer():
    def __init__(self, buffer_limit = 50000, min_weight=0.0001):
        self.buffer = deque(maxlen=buffer_limit)
        self.min_weight = min_weight
    
    def put(self, transitions, weights):
        assert len(transitions) == len(weights)
        self.buffer.extend(zip(transitions, weights))
    
    def sample(self, n):
        weights = [x[1] + self.min_weight for x in self.buffer]
        samples = random.choices(self.buffer, weights=weights, k=n)
        return [x[0] for x in samples]
    
    def __len__(self):
        return len(self.buffer)
```

```python
class Qnet(nn.Module):
    def __init__(self):
        super(Qnet, self).__init__()
        self.fc1 = nn.Linear(4, 32)
        self.fc2 = nn.Linear(32, 32)
        self.fc3_value = nn.Linear(32,1)
        self.fc3_advantage = nn.Linear(32, 2)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        
        advantages = self.fc3_advantage(x) # n x 2
        mean_advantage = advantages.mean(-1, keepdim=True)  # n x 1
        value = self.fc3_value(x)  # n x 1
        x = value + advantages - mean_advantage # n x 2
        return x
      
    def sample_action(self, obs, epsilon):
        coin = random.random()
        if coin < epsilon:
            return random.randint(0,1)
        else:
            out = self.forward(obs)
            return out.argmax().item()
```

```python
def plot_rewards(scores, epsilons):
    plt.plot(
        tuple(range(len(scores))),
        tuple(h for h in scores)
    )
    plt.draw()
    plt.show()

    plt.plot(
        tuple(range(len(epsilons))),
        tuple(h for h in epsilons)
    )
    plt.draw()
    plt.show()
```

```python
def run(q_net, epsilon, env):
    state = env.reset()
    transitions = []

    for i in range(1000):
        action = q_net.sample_action(torch.from_numpy(state).float(), epsilon)

        next_state, reward, done, info = env.step(action)
        running_mask = 0.0 if done else 1.0

        transitions.append((state, action, reward, next_state, running_mask))
            
        if done:
            return transitions
        else:
            state = next_state
    
    raise Exception("Infinite loop encountered!")
    
def train(q_net, q_target_net, memory, optimizer, gamma, batch_size=32):
    for _ in range(10):
        state, action, reward, next_state, running_mask = transitions_to_tensors(memory.sample(batch_size))

        q_out = q_net(state)
        
        q_action = q_out.gather(1, action)
        
        max_q_target_next = q_target_net(next_state).max(1)[0].unsqueeze(1)
        target = reward + gamma * max_q_target_next * running_mask
        
        loss = F.smooth_l1_loss(q_action, target)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
def main(q_net, init_epsilon = 0.5, target_epsilon = 0.001, n_iterations = 10000, learning_rate = 0.0005):
    assert target_epsilon <= init_epsilon

    env = gym.make('CartPole-v1')
    env.seed(RANDOM_SEED)
    
    replay_buffer = ReplayBuffer()    
    q_net = deepcopy(q_net)
    q_target_net = deepcopy(q_net)
    
    all_rewards = []
    all_epsilons = []
    
    avg_window = 50
    avg_rewards = []
    avg_epsilons = []
    
    gamma = 0.98
    epsilon = init_epsilon
    decay = math.e ** (math.log(target_epsilon/epsilon) / (n_iterations - 1))
    
    optimizer = optim.Adam(q_net.parameters(), lr=learning_rate)
    
    def _transition_weights(transitions):
        state, action, reward, next_state, running_mask = transitions_to_tensors(transitions)
        q_out = q_net(state)
        q_action = q_out.gather(1, action)
        max_q_target_next = q_target_net(next_state).max(1)[0].unsqueeze(1)
        target = reward + gamma * max_q_target_next * running_mask
        q_errors = q_action - target
        return q_errors.abs().squeeze().tolist()

    for i in range(n_iterations):
        transitions = run(q_net, epsilon, env)
            
        replay_buffer.put(transitions, _transition_weights(transitions))
        
        reward = len(transitions)
        all_rewards.append(reward)
        all_epsilons.append(epsilon)
        
        if len(replay_buffer) > 500:
            train(q_net, q_target_net, replay_buffer, optimizer, gamma)
            
            if i % 20 == 0:
                q_target_net = deepcopy(q_net)
        
        epsilon *= decay

        if len(all_rewards) >= avg_window:
            avg_reward = np.average(all_rewards[i-avg_window+1:i+1])
            avg_epsilon = np.average(all_epsilons[i-avg_window+1:i+1])
            
            if avg_reward >= 499 and epsilon < 0.05:
                break
            
            avg_rewards.append(avg_reward)
            avg_epsilons.append(avg_epsilon)
            
            if i % 1000 == 0:
                print(i, avg_reward, avg_epsilon)
    
    print(i, avg_reward, avg_epsilon)
    print(f"epsilon={epsilon}")
    print(f"q={q_net}")
    
    env.close()

    return q_net, all_rewards, avg_rewards, avg_epsilons
```

```python
q_net_0 = Qnet()
q_net_1, all_rewards, avg_rewards, avg_epsilons = main(q_net_0)
plot_rewards(avg_rewards, avg_epsilons)
```

```python
def display_frames_as_video(frames):
    fig = plt.figure(figsize=(frames[0].shape[1]/72.0, frames[0].shape[0]/72.0), dpi=72)
    plt.axis('off')
    
    ims = []
    for i, frame in enumerate(frames):
        im = plt.imshow(frame, animated=True)
        ims.append([im])
        
    anim = animation.ArtistAnimation(fig, ims, interval=40, blit=True, repeat=False)
    display(HTML(anim.to_html5_video()))
    plt.close()

def test(q):
    max_step = 1000
    
    env = gym.make('CartPole-v1')
    env.seed(RANDOM_SEED)
    env._max_episode_steps = max_step
    
    state = env.reset()
    frames = []

    for i in range(max_step):
        frame = env.render(mode="rgb_array")
        frames.append(frame)
        action = q.sample_action(torch.from_numpy(state).float(), 0)
        state, reward, done, info = env.step(action)
        if done:
            print(i)
            break
    env.close()
    display_frames_as_video(frames)
```
